```python
import numpy as np

# Inputs
x1, x2 = 0.6, 0.1
y = 1

# Parameters
w1, w2, b = 0.2, -0.3, 0.4
lr = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Forward pass
z = w1*x1 + w2*x2 + b
y_pred = sigmoid(z)

# Loss gradient (Binary Cross Entropy + Sigmoid shortcut)
dL_dy_pred = y_pred - y
dy_pred_dz = y_pred * (1 - y_pred)
dL_dz = dL_dy_pred * dy_pred_dz

# Gradients
dL_dw1 = dL_dz * x1
dL_dw2 = dL_dz * x2
dL_db  = dL_dz

# Update
w1 -= lr * dL_dw1
w2 -= lr * dL_dw2
b  -= lr * dL_db

print(w1, w2, b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# AND Gate Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training function
def train_model(learning_rate, epochs=100):
    w = np.array([0.1, 0.2])
    b = 0.3
    errors = []

    for epoch in range(epochs):
        total_error = 0

        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            y_pred = sigmoid(z)

            # Binary Cross Entropy Loss
            loss = -(y[i] * np.log(y_pred + 1e-9) +
                     (1 - y[i]) * np.log(1 - y_pred + 1e-9))

            total_error += loss

            # Gradient
            dz = y_pred - y[i]
            dw = dz * X[i]
            db = dz

            # Update weights
            w -= learning_rate * dw
            b -= learning_rate * db

        errors.append(total_error)

    return errors


# ---- ONLY MODIFICATION HERE ----
lr = 1.9  # change learning rate here
errors = train_model(lr)

plt.plot(errors)
plt.xlabel("Epoch")
plt.ylabel("Total Error")
plt.title(f"Error vs Epoch (LR = {lr})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# AND Gate Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training function
def train_model(learning_rate, epochs=100):
    w = np.array([0.1, 0.2])
    b = 0.3
    errors = []

    for epoch in range(epochs):
        total_error = 0

        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            y_pred = sigmoid(z)

            # Binary Cross Entropy Loss
            loss = -(y[i] * np.log(y_pred + 1e-9) +
                     (1 - y[i]) * np.log(1 - y_pred + 1e-9))

            total_error += loss

            # Gradients
            dz = y_pred - y[i]
            dw = dz * X[i]
            db = dz

            # Update
            w -= learning_rate * dw
            b -= learning_rate * db

        errors.append(total_error)

    return errors


# ---- TRY DIFFERENT LEARNING RATES ----
learning_rates = [0.01, 0.1, 0.5, 1.0, 1.9]

plt.figure()

for lr in learning_rates:
    errors = train_model(lr)
    plt.plot(errors, label=f"LR = {lr}")

plt.xlabel("Epoch")
plt.ylabel("Total Error")
plt.title("Error vs Epoch for Different Learning Rates")
plt.legend()
plt.show()
```
